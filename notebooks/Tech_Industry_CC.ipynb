```python
# import packages for alpaca, panadas and OS/dotenv to read .env file with api keys
import os
import pandas as pd
import numpy as np
import alpaca_trade_api as tradeapi
# Load .env enviroment variables
from dotenv import load_dotenv
load_dotenv()
import pathlib as Path
```

```python
#helpful maybe?
from datetime import datetime, timedelta

days = 90
startdate = datetime.now() - timedelta(days=days)

# for fixed dates
start = pd.to_datetime('2002-01-01')
end = pd.to_datetime('2020-01-01')
```

```python
# storing alpaca key and secret key in variables 
alpaca_api_key = os.getenv('ALPACA_API_KEY')
alpaca_secret_key = os.getenv('ALPACA_SECRET_KEY')

api = tradeapi.REST(alpaca_api_key,alpaca_secret_key,api_version='V2')
```

```python
# set time frame
timeframe = '1D'

# start and 'end' date for covid stock market 


start_date_covid = pd.Timestamp('2019-12-01',tz='America/New_York').isoformat()
end_date_covid = pd.Timestamp('2020-11-20',tz='America/New_York').isoformat()

# start and end date for 2008 recession 
start_date_2008 = pd.Timestamp('2007-07-01',tz='America/New_York').isoformat()
end_date_2008 = pd.Timestamp('2009-07-01',tz='America/New_York').isoformat()

# set tickers (per industry)
tech_tickers = ['AMD', 'AAPL', 'MSFT', 'GOOG', 'INTC']

# get stock data from covid
covid_tech_ticker = api.get_barset(tech_tickers,timeframe,start=start_date_covid,end = end_date_covid).df

# get stock data from 2008 recession
tech_ticker2008 = api.get_barset(tech_tickers,timeframe,start=start_date_2008,end = end_date_2008).df
```

```python
covid_tech_ticker.tail()
```

```python
# create new covid and 2008 dataframe to store ckosing prices of each stock

covid_tech_closingprices = pd.DataFrame()

# get closing prices for all tickers
for ticker in tech_tickers:
    covid_tech_closingprices[ticker]=covid_tech_ticker[ticker]['close']
    
# drop time component on index
covid_tech_closingprices.index = covid_tech_closingprices.index.date
    
covid_tech_closingprices.head()
```

```python
tech_closingprices2008 = pd.DataFrame()

# get closing prices for all tickers
for ticker in tech_tickers:
    tech_closingprices2008[ticker]=tech_ticker2008[ticker]['close']
    
# drop time component on index
tech_closingprices2008.index = tech_closingprices2008.index.date
    
tech_closingprices2008.head()
```

```python
# calculate daily returns for each timeframe
tech_covid_dr = covid_tech_closingprices.pct_change().dropna()
tech_returns2008 = tech_closingprices2008.pct_change().dropna()
```
